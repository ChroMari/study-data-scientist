## Условие

Яша плавал в бассейне размером N × M метров и устал. В этот момент он обнаружил, что находится на расстоянии x метров от одного из длинных бортиков (не обязательно от ближайшего) и y метров от одного из коротких бортиков. Какое минимальное расстояние должен проплыть Яша, чтобы выбраться из бассейна на бортик? Программа получает на вход числа N, M, x, y. Программа должна вывести число метров, которое нужно проплыть Яше до бортика.

```python
n = int(input('Введите размер бассейна по стороне N: '))
m = int(input('Введите размер бассейна по стороне M: '))
x = int(input('Введите расстояние на котором находится Яша от стороны бассейна: '))
y = int(input('Введите расстояние на котором находится Яша от другой стороны бассейна: '))

# сначала распределяем стооны. Находим большую и меньшую
if n > m:
    n, m = m, n
    
if x >= n / 2:
    x = n - x
if y >= n / 2:
    y = m - y

if x < y:
    print(f'Яше нужно проплыть {x}')
else:
    print(f'Яше нужно проплыть {y}')
```
