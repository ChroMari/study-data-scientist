### DataFrame.describe()

Генерация описательной статистики.

К описательной статистике относятся те, которые суммируют центральную тенденцию, дисперсию и форму распределения набора данных, исключая NaNзначения.

Анализирует как числовые, так и ряды объектов, а также DataFrameнаборы столбцов смешанных типов данных. Результат будет варьироваться в зависимости от того, что предоставляется. 

```python
students_performance.describe()
```

### DataFrame.groupby(значение)

*Обязательно нужно указывать хотя бы одну колонку по которой будет происходить группировка.*

Группируйте DataFrame с помощью преобразователя или серии столбцов.

Операция группировки включает в себя некоторую комбинацию разделения объекта, применения функции и объединения результатов. Это можно использовать для группировки больших объемов данных и выполнения операций над этими группами.

### DataFrameGroupBy.aggregate(...)

Агрегация с использованием одной или нескольких операций по указанной оси.

```python
students_performance.groupby('gender').aggregate({'writing score': 'mean'})
```

```python
students_performance.size
```

### DataFrame.iloc[ ... ]

Чисто целочисленная индексация на основе местоположения для выбора по позиции.

.iloc[]в основном основан на целочисленной позиции (от оси 0до ), но также может использоваться с логическим массивом.length-1

- Первый срез относится к строкам, которые будут взяты из данных
- Второй срез относится к колонкам, которые будут взяты из данных

Можно указывать не срезы, а конкретные строки или колонки, которые нам нужны. Для этого вместо среза будет использоваться синтаксис [1, 5, 89].

```python
students_performance.iloc[0:3, 0:5]
```

```python
students_performance_with_names = students_performance.iloc[[0, 3, 4, 7, 8]]
```

```python
students_performance_with_names
```

### DataFrame.index = [ ... ]

Индекс (метки строк) DataFrame.

Индекс DataFrame — это серия меток, идентифицирующих каждую строку. Метки могут быть целыми числами, строками или любым другим хешируемым типом. Индекс используется для доступа и выравнивания на основе меток, и к нему можно получить доступ или изменить его с помощью этого атрибута.

```python
students_performance_with_names.index = ['Cersei', 'Tywin', 'Gregor', 'Joffrey', 'Ilyn Payne']

students_performance_with_names
```

### DataFrame.loc[[...]]

Доступ к группе строк и столбцов по меткам или логическому массиву.

.loc[]в основном основан на метках, но может также использоваться с логическим массивом.

Сначала указываются строки, вторым списком можно было бы указать также какие именно колонки нам нужно взять. Если колонки не указываются, то по умолчанию берутся все.

*Если индексы идут индексевые, то мы можем их указывать. Также можно использовать срезы для получения данных. Но нужно учитывать, что браться индексы будут включительно по указанные позиции.*

```python
students_performance_with_names.loc[['Cersei', 'Joffrey']]
```
