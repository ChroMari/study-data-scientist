```python
import pandas as pd
import numpy as np
```

```python
students_performance = pd.read_csv('./StudentsPerformance.csv')
```

```python
students_performance = students_performance.rename(columns=lambda x: x.replace(' ', '_'))
```

```python
students_performance.head()
```

```python
students_performance.math_score.unique() #выводит все угикальные значения колонок
```

```python
students_performance.groupby('gender').mean(numeric_only=True)
```

```python
students_performance.groupby('gender', as_index=False).aggregate({'math_score':'mean','reading_score':'mean','writing_score':'mean'})
```

```python
students_performance.groupby('gender').aggregate({'math_score':'mean'})
```

```python
students_performance.groupby(['gender', 'lunch']).mean(numeric_only=True)
```

```python
students_performance.groupby(['gender', 'lunch']).math_score.nunique()
```

```python
students_performance \
    .sort_values(['gender', 'math_score'], ascending=False) \
    .groupby('gender') \
    .head()
```

```python
students_performance['total_score'] = students_performance.math_score + students_performance.reading_score + students_performance.writing_score
```

### DataFrame.agg({ ... })

Агрегация с использованием одной или нескольких операций по указанной оси.

Слевой стороны мы указываем колонку, а с правой стороны указываем метод, который нам нужно применить для этой колонки.

```python
dota.groupby('legs').agg({'legs': 'count'})
```

```python
dota.groupby(['attack_type', 'primary_attr']).count()
```
